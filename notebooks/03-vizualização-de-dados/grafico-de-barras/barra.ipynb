```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

### Visualização #01

**Gráfico de barras**

```python
df1 = pd.read_csv('/content/gist.githubusercontent.com_omadson_770934cb0df9e2d2c3e8db031f1139a5_raw_4d181efc035b327d9d66256915951384517c603e_arrecadacao.csv')
```

```python
df1
```

```python
df1 = df1.assign(arrecadacao_milhao = df1.arrecadacao / 1000000)
df1 = df1.sort_values(by= 'arrecadacao', ascending= False)

#add linhas verticais
sns.set_style("whitegrid")

# horizontal e uma única cor
plt.figure(figsize=(5, 2))
ax = sns.barplot(data=df1, x="arrecadacao_milhao", y = 'titulo', color= "tomato")

#add valores nas barras
ax.bar_label(
    ax.containers[0],
    fmt = '%.1f',
    label_type = 'center',
    color = 'white'
)

# labels descritivos
plt.xlabel('Arrecadação(Milhões de Dólares)')
plt.ylabel('Filme')
plt.title('Arrecadação de Filmes')

#linha vertical na origem e remoção das bordas
ax.set(frame_on= False)
ax.axvline(0, color='k', clip_on=False)

plt.show()
```
