* **matplotlib** : biblioteca de visualização de dados (médio nível)
* **seaborn** : biblioteca de visualização de dados (alto nível)

```python
import pandas as ps
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = sns.load_dataset('tips')
df.head(5)
```

### Visualização #03


**Boxplot**

```python
sns.set_style('whitegrid')
plt.figure(figsize=(6, 3))
sns.boxplot(x='day', y='total_bill', data=df, color="cornflowerblue")

#titulo e labels
plt.xlabel('Dia da Semana')
plt.ylabel('Conta(USD)')
plt.title('Distriuição de gorjetas ao longo da semana')
plt.show()
```
