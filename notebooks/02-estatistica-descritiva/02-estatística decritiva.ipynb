### **Quarteto anscombe**

```python
df4 = sns.load_dataset('anscombe')
```

```python
df4
```

Dicionário:

O conjunto de dados "Anscombe" no seaborn é um exemplo clássico na estatística, criado por Francis Anscombe em 1973. Ele é composto por quatro conjuntos distintos de dados que possuem as mesmas propriedades estatísticas (média, variância, correlação, e equação de regressão), mas aparecem muito diferentes quando visualizados graficamente.

A ideia por trás desse conjunto de dados é demonstrar a importância da visualização de dados e como estatísticas simples podem ser insuficientes para entender um conjunto de dados.

As colunas no conjunto de dados "Anscombe" são:

**dataset** - Indica a qual dos quatro conjuntos (I, II, III, ou IV) o ponto pertence. Cada um desses conjuntos tem sua própria particularidade quando visualizado em um gráfico.

**x** - O valor da variável independente.

**y** - O valor da variável dependente.

```python
df4_dict = pd.DataFrame([
    {
        'variavel': 'dataset',
        'descricao': 'Idica a qual conjunto de dados os valores pertencem',
        'tipo': 'Qualitativa',
        'subtipo': 'nominal',
        'resposta': False
    },{
        'variavel': 'x',
        'descricao': 'Variavel independente',
        'tipo': 'Quantitativa',
        'subtipo': 'discreta',
        'resposta': False
    },{
        'variavel': 'y',
        'descricao': 'Variavel dependente',
        'tipo': 'Quantitativa',
        'subtipo': 'contínua',
        'resposta': True
    }
])
df4_dict
```

```python
g = sns.FacetGrid(df4, col="dataset", col_wrap=2, height=4)
g.map(sns.scatterplot, "x", "y", alpha=0.7)
g.add_legend()

plt.show()
```
